# Example: using mie_scattering for lognomal aerosol
This script will use mie_scattering to compute the scattering properties of an aerosol particles by averaging over a lognormal size distribution. 

```python
%pylab inline
```

```python
# Standard

# 3rd parth
import scipy as sp
import matplotlib.pyplot as plt
# import the lognormal 
import mie_scattering as mie
```

```python
# Define a size distribution 
wavelength = .532
reff = 0.1
veff = 0.05
mr = 1.45
mi = 0.001

# Calculate scattering properties of this particle distribution 
ssa = mie.lognorm(wavelength, mr, mi, reff, veff)
```

```python
# The volume-scattering and volume extinction coeffients
ext = ssa.ext()
sca = ssa.sca()
volume_absorption_coeff = ext - sca
single_scattering_albedo = sca / ext

# Calculate the angular variability 
theta = sp.linspace(0, sp.pi, 180)
out = ssa.angle_eval(theta)
```

```python
# Make a figure with the scattering properties
fig = plt.figure(0, figsize=(10,10), facecolor='white')
ax = fig.add_axes([.1, .1, .85, .85])
ax.plot(theta, out.transpose(), linewidth=4)
```
